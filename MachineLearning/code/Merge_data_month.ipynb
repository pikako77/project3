```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
file = '../data/Total_consumption_Coal_gas_petroleum_Texas_monthly.csv'
filePop = '../data/population_out_yr.csv'

out_file = '../data/Total_consumption_Coal_gas_petroleum_population_Texas_monthly.csv'
```

```python
data = pd.read_csv(file)
data.head()
```

```python
len(data)
```

```python
dataPop = pd.read_csv(filePop)
dataPop.head()
```

```python
# Define interpolation function
def interpolation(x1,x2,y1,y2, x):
    y = y1 + (y2 - y1)/(x2-x1) * (x - x1)
    return y


interpY = interpolation(2011,2012,0,10, 2011.1)
print(interpY)
```

```python
yr_interp = []
mo_interp = []

pop_yr =  dataPop['year']

step = 1/12

start_yr = 2001
end_yr  =  2019

yr = start_yr

tmp_interp = []

while yr < end_yr:
    yr1 = yr
    yr2 = yr + 1
    
    idx1 = dataPop[dataPop['year'] == yr1].index
    idx2 = dataPop[dataPop['year'] == yr2].index
    pop1 = dataPop['population'].iloc[idx1].values[0]
    pop2 = dataPop['population'].iloc[idx2].values[0]    
    
    #print (yr1, pop1)
    #print(yr2, pop2)
    for i in range (12):
        tmp =  interpolation(yr1, yr2, pop1, pop2, yr1+i*step)
        tmp_interp.append(tmp)
        yr_interp.append(yr1)
        mo_interp.append(i+1)
        print(yr1 ,i+1, tmp)

    yr = yr2
    
yr_interp.append(2019)
mo_interp.append(1)
tmp_interp.append(dataPop[dataPop['year'] == 2019]['population'].values[0])

```

```python
pop_interp_mo = pd.DataFrame ( {
    'year': yr_interp, 
    'month':mo_interp, 
    'population': tmp_interp
})

pop_interp_mo
```

```python
pop_interp_mo.to_csv('pop_monthly.csv')
```

```python
pd.concat([data,pop_interp_mo])
```

```python
merge_data = pd.merge(data , pop_interp_mo, right_on='year', left_on = 'year', how = 'inner')
len(merge_data)
```

```python
merge_data.dropna()
merge_data.drop_duplicates()
```

```python
merge_data.to_csv(out_file,index=False)
```
