#### Interpolation function

```python
# Define interpolation function
def interpolation(x1,x2,y1,y2, x):
    y = y1 + (y2 - y1)/(x2-x1) * (x - x1)
    return y


interpY = interpolation(1910,1920,0,10, 1911)
print(interpY)
```

```python
yr_interp = []

start_yr = 1960
end_yr = 2017
step = 1
yr  = start_yr

yr_interp.append(yr) 
while yr < end_yr:
    yr = yr + step
    yr_interp.append(yr)

len(yr_interp)
```

```python
interp_pop = []
index = []

for i in range(len(yr_b4)-1):
    if (float(yr_b4[i+1]) - float(yr_b4[i])) > 1:
    #print(yr_b4[i],yr_b4[i+1])
        for yr_interval in yr_interp: 
            if yr_interval >= float(yr_b4[i]) and yr_interval < float(yr_b4[i+1]):
                    interpol_tmp = interpolation(float(yr_b4[i]),float(yr_b4[i+1]),pop_b4[i],pop_b4[i+1], yr_interval)
                    #print(yr_interval, interpol_tmp)
                    interp_pop.append(interpol_tmp)
                    index.append(i)
        
len(interp_pop)
```
